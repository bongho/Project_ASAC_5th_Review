```python
# 3개의 단어 묶음인 단어구 추출
n_gram_range = (3, 3)
stop_words = "english"

count = CountVectorizer(ngram_range=n_gram_range, stop_words=stop_words).fit([doc])
candidates = count.get_feature_names_out()

print('trigram 개수 :',len(candidates))
print('trigram 다섯개만 출력 :',candidates[:5])
```

```python
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
doc_embedding = model.encode([doc])
candidate_embeddings = model.encode(candidates)
```

```python
# 코사인 유사도 계산
distances = cosine_similarity(doc_embedding, candidate_embeddings)

# 상위 N개의 키워드 선택
top_n = 5
top_n_indices = distances.argsort()[0][-top_n:]

# 키워드와 유사도 퍼센트 출력
keywords_with_scores = [(candidates[index], distances[0][index] * 100) for index in top_n_indices]
keywords_with_scores = sorted(keywords_with_scores, key=lambda x: x[1], reverse=True)  # 유사도 순으로 정렬
```

```python
# 결과 출력
print("Top {} keywords with similarity scores:".format(top_n))
for keyword, score in keywords_with_scores:
    print("-------------"*5)
    print(f"Keyword: {keyword}, Similarity: {score:.2f}%")
```
