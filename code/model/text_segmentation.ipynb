```python
import nltk
from nltk.tokenize import PunktSentenceTokenizer
from gensim import corpora, models
import gensim
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
# 문장으로 분리
tokenizer = PunktSentenceTokenizer()
sentences = tokenizer.tokenize(doc)
```

```python
sentences
```

```python
# 텍스트 전처리 함수
def preprocess_text(document):
    # 소문자로 변환
    document = document.lower()
    # 문장 토크나이징
    words = nltk.word_tokenize(document)
    # 불용어 제거 및 스테밍
    # words = [word for word in words if word not in stopwords.words('english')]
    # stemmer = PorterStemmer()
    # words = [stemmer.stem(word) for word in words]
    return words
```

```python
# 각 문장을 전처리
texts = [preprocess_text(sentence) for sentence in sentences]

# 사전 생성
dictionary = corpora.Dictionary(texts)

# 문장별로 Corpus 생성
corpus = [dictionary.doc2bow(text) for text in texts]

# LDA 모델 생성 및 훈련
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=3, id2word = dictionary, passes=15)
```

```python
texts
```

```python
# 각 문장을 가장 높은 확률의 주제에 할당하고 주제별 문장 저장
topic_sentences = {i: [] for i in range(ldamodel.num_topics)}
for i, row in enumerate(ldamodel[corpus]):
    row = sorted(row, key=lambda x: (x[1]), reverse=True)
    topic_sentences[row[0][0]].append(sentences[i])

# 각 토픽에 대한 주제와 문장 합친 것을 출력
for topic_num, sentences in topic_sentences.items():
    topic_details = ldamodel.show_topic(topic_num, topn=5)  # 주제별 주요 단어 5개
    topic_keywords = ", ".join([word for word, prop in topic_details])
    print(f"토픽{topic_num}: {' '.join(sentences)},   토픽주제: {topic_keywords}")
```

```python
import nltk

# wordnet 다운로드
nltk.download('wordnet')
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer

# 전처리
tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(document):
    # 토크나이징과 소문자 변환
    tokens = tokenizer.tokenize(document.lower())
    # 불용어 제거 및 lemmatization
    cleaned_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    return cleaned_tokens

texts = [preprocess_text(doc)]

# 사전 및 코퍼스 생성
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# LDA 모델 생성
ldamodel = gensim.models.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=20)

# 결과 출력
topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```
