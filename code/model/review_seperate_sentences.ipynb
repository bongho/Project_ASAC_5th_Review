# 텍스트 전처리

> \n 제거, (?) 제거

```python
# \n, (?), (?, ?)을 공백으로 대체하는 코드
review_df['text'] = review_df['text'].str.replace(r'\n', ' ', regex=True)
review_df['text'] = review_df['text'].str.replace(r'\(\?\)', ' ', regex=True)
review_df['text'] = review_df['text'].str.replace(r'\(\?, \?\)', ' ', regex=True)
```

# 데이터 프레임에 적용하여 문장수 컬럼 추가

```python
# 문장 개수 계산 및 데이터프레임에 추가
review_df['text_length'] = review_df['text'].apply(lambda x: len(tokenizer.tokenize(x)))
```

```python
# '(?)'이 포함된 행을 찾는 코드
contains_question_mark = review_df['text'].str.contains(r'\(\?\)')
rows_with_question_mark = review_df[contains_question_mark]

# 해당 행을 출력
rows_with_question_mark['text']
```

## text_length 분포 확인

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis
```

```python
# 히스토그램
plt.figure(figsize=(10, 6))
ax = sns.histplot(review_df['text_length'], bins=10, kde=False, color='skyblue')
# 각 막대 위에 수치 추가
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.title('Histogram of review_df with Counts')
plt.xlabel('review_df')
plt.ylabel('count')
plt.show()

# 박스 플롯
plt.figure(figsize=(10, 6))
sns.boxplot(x=review_df['text_length'], color='skyblue')
plt.title('Box Plot of text_length')
plt.xlabel('text_length')
plt.show()

# 커널 밀도 추정 (KDE) 플롯
plt.figure(figsize=(10, 6))
sns.kdeplot(review_df['text_length'], shade=True, color='skyblue')
plt.title('KDE Plot of text_length')
plt.xlabel('text_length')
plt.ylabel('Density')
plt.show()
```

```python
pd.options.display.float_format = '{:.0f}'.format
review_df['text_length'].describe()
```

```python
# 왜도와 첨도
skewness = review_df['text_length'].skew()
kurtosis = review_df['text_length'].kurt()
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

```python

# 히스토그램과 KDE 플롯
plt.figure(figsize=(10, 6))
ax = sns.histplot(review_df['text_length'], bins=30, kde=True, color='skyblue')

# 각 막대 위에 수치 추가
for p in ax.patches:
    height = int(p.get_height())
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='baseline', fontsize=8, color='black', xytext=(0, 5),
                textcoords='offset points')

# 왜도와 첨도 계산
skewness = skew(review_df['text_length'])
kurt = kurtosis(review_df['text_length'])

# 왜도와 첨도 텍스트로 추가
plt.title(f'Histogram of Text Length with Counts\nSkewness: {skewness:.2f}, Kurtosis: {kurt:.2f}')
plt.xlabel('Text Length')
plt.ylabel('Count')
plt.show()

# 왜도와 첨도 시각화를 위한 KDE 플롯
plt.figure(figsize=(10, 6))
sns.kdeplot(review_df['text_length'], shade=True, color='skyblue')
plt.axvline(review_df['text_length'].mean(), color='r', linestyle='--', label=f'Mean: {review_df["text_length"].mean():.2f}')
plt.title(f'KDE Plot of Text Length\nSkewness: {skewness:.2f}, Kurtosis: {kurt:.2f}')
plt.xlabel('Text Length')
plt.ylabel('Density')
plt.legend()
plt.show()
```
