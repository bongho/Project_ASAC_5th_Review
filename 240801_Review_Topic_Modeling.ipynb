## 데이터 불러오기

```python
review_df = pl.read_ndjson("./data/yelp_academic_dataset_review.json")
print(review_df.shape)
review_df.head()
```

```python
review_df = review_df.to_pandas()
review_df = review_df.sample(n=100000)
print(review_df.shape)
review_df.head()
```

## 리뷰 데이터 전처리 -> 특정 가게로 한정하기

```python
## 특정 가게 중 리뷰가 많은 가게를 하나 임의 선택
sample_business_id = review_df['business_id'].value_counts().reset_index().iloc[0,0]
print("sample_business_id: ", sample_business_id)
```

```python
sample_df = review_df[review_df['business_id'] == sample_business_id]
print(sample_df.shape)
sample_df.head()
```

## 리뷰 데이터 전처리 -> Text 데이터 전처리

```python
#Sample Df의 Text Column을 "."를 기준으로 나눈다음, review_id,"."로 나눠진 text로 이루어진 DataFrame을 만든다. 
explode_df  = sample_df['text'].str.split(".").explode()
explode_df  = sample_df[['review_id']].join(explode_df)
explode_df = explode_df[explode_df['text'] != ""] # 빈 문자열 제거
explode_df['text'] = explode_df['text'].str.replace("\n", " ").str.replace("\t", " ").str.replace("\r", " ")
print(explode_df.shape)
explode_df.head()
```

## Bertopic을 이용한 Topic Modeling

```python
!pip install bertopic
```

```python
from bertopic import BERTopic
```

```python
docs = explode_df['text'].tolist()
docs[:2]
```

```python
topic_model = BERTopic(language="english", nr_topics=3 ,calculate_probabilities=True, verbose=True) #4로 한 것은 outlier topic을 포함한 것이다.
topics, probs = topic_model.fit_transform(docs)
```

```python
len(probs),len(docs), topic_model.get_topic_info()['Count'].sum()
```

```python
topic_model.get_topic_info()
```

```python
# topics를 explode_df에 추가
explode_df['topic'] = topics
explode_df.head(30)
```

## 리뷰 별 토픽 리스트업, -1은 제거하였음

```python
# topic_df = explode_df.groupby('review_id')['topic'].apply(lambda x: list(set(x.tolist()))).reset_index()
topic_df = explode_df.groupby('review_id')['topic'].apply(lambda x: [item for item in set(x.tolist()) if item != -1]).reset_index()

# 이 때 topic 컬럼의 리스트 값을 오름차순으로 정렬한다.
topic_df['topic'] = topic_df['topic'].apply(lambda x: sorted(x))
print(topic_df.shape)
topic_df.head()

```

```python
import pandas as pd
topic_df = pd.DataFrame(topic_df['topic'].values.tolist()).add_prefix('topic_').join(topic_df)
print(topic_df.shape)
topic_df.head()
```

## 리뷰별 특정 토픽 여부를 컬럼으로 분류하여 표기

```python
import numpy as np
# topic_df의 topic_0 ~ topic_2까지 확인해서 nan이면 0 아니면 1로 변경하기
topic_df.iloc[:, :2] = np.where(topic_df.iloc[:, :2].isnull(), 0, 1)
```
